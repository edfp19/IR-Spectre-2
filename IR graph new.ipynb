```python
import pandas as pd
import math
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style


def transf_t_abs(df):
    df["abs"]=df["abs"]/100
    df["abs"]=df["abs"].apply(lambda x: 2 - math.log(x))
    df["abs"]=df["abs"]/df["abs"].max()
    return df

P_80miel = pd.read_csv("FILM PECT+ 80 MIEL.CSV")

P_80miel = transf_t_abs(P_80miel)


Pectina_60_5beep = pd.read_csv("films pectina+60 miel+5beep.CSV")
Pectina_60_5beep = transf_t_abs(Pectina_60_5beep)

Pectina_100 = pd.read_csv("films pectina+100 miel.CSV")
Pectina_100 = transf_t_abs(Pectina_100)

miel_apis_abs = pd.read_csv("Miel Apis.CSV")
miel_apis_abs = transf_t_abs(miel_apis_abs)

PC_40_Gly = pd.read_csv("PC+40 GLY.CSV")
PC_40_Gly = transf_t_abs(PC_40_Gly)

Pec_Alt_Met = pd.read_csv("Pectina de alto metoxilo.CSV")
Pec_Alt_Met= transf_t_abs(Pec_Alt_Met)

agua_csv = pd.read_csv("Agua.CSV")
agua_csv= transf_t_abs(agua_csv)

P_60miel = pd.read_csv("FILM PECT+ 60 MIEL.CSV")
P_60miel= transf_t_abs(P_60miel)




fig, ax= plt.subplots(figsize=(13,6))
ax.plot(P_80miel["long_onda"],P_80miel["abs"], label="P80_miel")
ax.plot(agua_csv["long_onda"], agua_csv["abs"], label="Agua")
ax.plot(Pectina_60_5beep["long_onda"], Pectina_60_5beep["abs"], label="Pectina + 60 + 5beep")
ax.plot(Pec_Alt_Met["long_onda"], Pec_Alt_Met["abs"], label="Pectina de alto metxido")
ax.plot(PC_40_Gly["long_onda"], PC_40_Gly["abs"], label="Pectina + 40GLY")
ax.plot(Pectina_100["long_onda"], Pectina_100["abs"], label="Pectina + 100% miel")
ax.plot(miel_apis_abs["long_onda"], miel_apis_abs["abs"], label="Miel de Apis")
ax.plot(P_60miel["long_onda"], P_60miel["abs"], label="Pectina + 60% miel")
#cutting the plot 
ax.legend(fontsize=8, loc="lower right", borderaxespad=1)
ax.set_xlim(4000, 500)
ax.set_ylim(None, 1.05)
ax.set_xlabel("Longitud de onda")
ax.set_ylabel("Absorbancia")
ax.set_title("Espectro IR")
ax.grid(True)
plt.tight_layout()
#plt.subplots_adjust(left=0.5, right=1, top=0.9, bottom=0.1)
plt.savefig("IR_Graph_New_LottaAdata.png")

plt.show()

```
